```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
import seaborn as sns
from numpy import polyfit, poly1d
```

```python
# Sampling rate 1000 hz / second
t = np.linspace(0, 2, 1000, endpoint=True)
# Plot the square wave signal
plt.plot(t, signal.square(3 * np.pi * 1 * t)/2 + 0.5, color='r')
# Give a title for the square wave plot
#plot.title('Sqaure wave - 5 Hz sampled at 1000 Hz /second')
# Give x axis label for the square wave plot
plt.xlabel('Time')
# Give y axis label for the square wave plot
plt.ylabel('Voltage')
# plot.grid(True, which='both')
# Provide x axis and line color
 # Set the max and min values for y axis
plt.ylim(0, 1.2)
plt.xlim(0,2)

plt.gca().axes.get_yaxis().set_ticklabels([])
plt.gca().axes.get_xaxis().set_ticklabels([])
# plt.gca().axes.get_yaxis().set_visible(False)
# plt.gca().axes.get_xaxis().set_visible(False)


# Display the square wave drawn
plt.show()
```

```python
array = [[12,0,0,0,0,0],
           [0,12,0,1,0,0],
           [0,0,12,0,0,0],
           [0,0,0,11,0,0],
           [0,0,0,0,12,0],
           [0,0,0,0,0,12]
           ]
col_array = ['D-T-V','D-V-T','T-D-V','T-V-D','V-T-D','V-D-T']

norm = np.linalg.norm(array)
normal_array = array/norm
print(normal_array)

per_array = []

for ele_array in array:
  per_ele_array = []
  for ele in ele_array:
    ele_b = ele / 12
    ele_b = round(ele_b, 2)
    per_ele_array.append(ele_b)
  per_array.append(per_ele_array)

print(per_array)


df_cm = pd.DataFrame(per_array, index=col_array, columns=col_array)
plt.figure(figsize=(10,7))
sns.set(font_scale=1.8) # for label size
# sn.heatmap(df_cm, annot=True, annot_kws={"size": 14}, linewidths=.1,cmap='YlGnBu') # font size
# plt.xlabel("True Order")
# plt.ylabel("Perceived Order")

ax = sns.heatmap(df_cm, annot=True, annot_kws={"size": 20}, linewidths=.1,cmap='YlGnBu') # font size
# plt.xlabel("True Order")
# plt.ylabel("Perceived Order")
ax.set(xlabel="True Order", ylabel="Perceived Order")
ax.xaxis.set_label_position('top')
# ax.tick_params(length=0)
plt.tick_params(axis='both', which='major', labelsize=12, labelbottom = False, bottom=False, top = False, labeltop=True)

plt.show()
```
